After the aforementioned code was executed, it was discovered that create columns contained some duplicate data in addition to case-sensitive data.

````python
print('Number of unique values in make column before cleaning :',len(makes))
````

````python
#changing all uppercase values to lower case values to reduce the number of unique values
df['make']=df['make'].str.lower()
````

````python
# type(df['make'][0])
len(df['make'].unique())
````

````python
# updating multiple formats or duplicate data with one unique value

import re

make_values = {'chev truck':'chevrolet','dodge tk':'dodge','ford tk':'ford','ford truck':'ford','gmc truck':'gmc',
              'hyundai tk':'hyundai','land rover':'landrover','mazda tk':'mazda','mercedes-b':'mercedes',
               'mercedes-benz':'mercedes','vw':'volkswagen'}
pattern= re.compile(r'\b(?:' + '|'.join(make_values.keys()) + r')\b')

def replace_make_values(make):
    return pattern.sub(lambda x: make_values[x.group()], make)


df['make']= pd.Series([replace_make_values(make) for make in df['make']])
````

````python
# final length of make column
print('Number of unique values in make column after cleaning :', len(df.make.unique()))
````

Sample narrative description.

1. Data that was normalized included several formats and case-sensitive data.
2. There are 97 distinct values in the make columns before cleaning.
3. There are 56 distinct values in the make column after cleaning.
